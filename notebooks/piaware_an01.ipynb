```python
# load a day's worth of PiAware data
import os
data_dir = '/data/piaware'
data_ts_ymdh = ['2021', '09', '19', '00']
data_path_day = os.sep.join([data_dir] + data_ts_ymd)
print(f'data_path_day: "{data_path_day}"')
print(data_path_day)
hourly_filenames = os.listdir(data_path_day)

# generate pathnames for an hour's worth of data
hourly_pathnames = []
for hourly_filename in hourly_filenames:
    #print(hourly_filename)
    hourly_pathnames.append(os.sep.join([data_path_day, hourly_filename]))
    
print()
#print('hourly_pathnames:')
#print(hourly_pathnames)
```

```python
# load single PiAware JSON file
path = hourly_pathnames[0]
print(path)
```

```python
import json
f = open(path)
pa_data = json.load(f)
f.close
```
